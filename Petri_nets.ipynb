# <center>Сети Петри</center>

## 1)

#### Классификация 

##### Динамические ограничения

- Сеть не является живой (активной), потому что существует такая маркировка (тупиковая), из которой все переходы будут заблокированы. (Невозможно вернуться в текущее состояние за конечное число переходов)


- Сеть является безопасной, потому что каждая позиция содержит не более одной метки


- Сеть является 1-ограниченной, так как максимальное число меток равно 1 (доказывает предыдущий пункт). Сеть является ограниченной, так как найдено такое целое k, для которого она k - ограничена.


- Сеть является консервативной, потому что сумма меток остаётся постоянной и равной 2


- Сеть является не устойчивой, так как срабатывание одного из возможных переходов снимает возбуждение с других переходов (если сработает переход t4, то с перехода t1 снимается возбуждение)


##### Статические ограничения
- Сеть свободного выбора, потому что каждая дуга, выходящая из позиции, является либо единственным выходом из нее, либо единственным входом в переход.


- Не является маркировочным графом, потому что позиция р4 имеет два выхода, а позиция р5 не имеет выходов.


- Не автоматная, т.к. переход t3 имеет более одного входа и выхода.


- Не бесконфликтная, т.к. позиция р4 является входной для t1 и t4, но не является для них выходной.



Для сети возможно параллельное срабатывание нескольких переходов

## 2)

#### Классификация 

##### Динамические ограничения

- Сеть является живой (активной), потому что не существует такой маркировки (тупиковой), из которой все переходы будут заблокированы. (Возможно вернуться в текущее состояние за конечное число переходов)


- Сеть не является безопасной, потому что позиция p6 содержит более одной метки 


- Сеть является 2-ограниченной, так как максимальное число меток равно 2 (подтверждает предыдущий пункт). Сеть является ограниченной, так как найдено такое целое k, для которого она k - ограничена.


- Сеть является неконсервативной, потому что сумма меток изменяется и может принимать значения 2, 4


- Сеть является устойчивой, так как срабатывание одного из возможных переходов не снимает возбуждение с других переходов.


##### Статические ограничения
- Сеть свободного выбора, потому что каждая дуга, выходящая из позиции, является либо единственным выходом из нее, либо единственным входом в переход.


- Маркировочным графом, потому что каждая позиция имеет в точности по одному входному и одному выходному переходу.


- Не автоматная,потому что переход t4 имеет более одного входа, а t1 имеет более одного выхода.


- Бесконфликтная, потому что для каждой позиции существует не более одной исходящей дуги.


Для сети возможно параллельное срабатывание нескольких переходов
